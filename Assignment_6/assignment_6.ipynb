```python
#Question 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Visualization 1

df=pd.read_csv('weather.csv')
data=df['MinTemp']
sns.distplot(data)
```

```python
#Visualization 2

sns.distplot(data,kde=False,rug=True)
```

```python
#Visualization 3

sns.distplot(data,hist=False,rug=True)
```

```python
#Visualization 4

sns.kdeplot(data,shade=True)
```

```python
#Visualization 5

data2=df['Humidity9am']
sns.kdeplot(data2, label='original')
sns.kdeplot(data2, bw=0.3, label='bw=0.3')
sns.kdeplot(data2, bw=2, label='bw=2')
plt.legend()
```

```python
#Visualization 6

df.plot(kind='scatter', x = 'Temp9am', y = 'Temp3pm')
```

```python
#Visualization 7

sns.jointplot(x='Pressure9am', y='Pressure3pm', data=df)
```

```python
#Visualization 8

x, y = df['Humidity9am'],df['Humidity3pm']
with sns.axes_style('white'):
    sns.jointplot(x = x, y = y, kind = 'hex', color = 'k')
```

```python
#Visualization 9

sns.jointplot(x='MaxTemp', y='MinTemp', data=df, kind='kde')
```

```python
#Visualization 10

sns.stripplot(x = 'WindDir9am', y='WindSpeed9am', data=df)
```

```python
#Visualization 11

sns.swarmplot(x='RainToday', y='Sunshine', data=df)
```
